```python
import matplotlib.pyplot as plt
import numpy as np
from utils.functions import PiecewiseLinear
```

```python

xs = [0, 1, 2, 4]
ys = [0, 2, 1, 3]
func = PiecewiseLinear(xs, ys)
# Rescale by factor of 1/2
rescaled = func.rescale_x(1/2)
# Create sample points for plotting
x_original = np.linspace(0, 4, 100)
x_rescaled = np.linspace(0, 2, 100)
# Calculate function values
y_original = [func(x) for x in x_original]
y_rescaled = [rescaled(x) for x in x_rescaled]
# Plot the functions
plt.figure(figsize=(10, 5))
plt.plot(x_original, y_original, 'b-', label='Original Function')
plt.plot(x_rescaled, y_rescaled, 'r--', label='Rescaled (n=2)')
# Add markers at the original control points
plt.plot(xs, ys, 'bo', markersize=8)
# Add markers at the rescaled control points
rescaled_xs = [x/2 for x in xs]
plt.plot(rescaled_xs, ys, 'ro', markersize=8)
plt.title('PiecewiseLinear Function Rescaling')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
