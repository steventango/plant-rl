```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_grey_squares(offsets):
    fig, ax = plt.subplots(figsize=(5, 5))  
    all_x = []
    all_y = []  
    for i, (offset_x, offset_y) in enumerate(offsets):
        square = patches.Rectangle((offset_x, offset_y), 1, 1,                
                 linewidth=0.5, edgecolor='k', facecolor='k', alpha=0.1)
        ax.add_patch(square)
        all_x.extend([offset_x, offset_x + 1])
        all_y.extend([offset_y, offset_y + 1])
    ax.set_xlim(min(all_x) - 0.2, max(all_x) + 0.2)
    ax.set_ylim(min(all_y) - 0.2, max(all_y) + 0.2)
    #ax.set_xlim(0,1)
    #ax.set_ylim(0,1)
    ax.set_aspect('equal')
```

```python
### Input tile coding config ###
tiles = 2
tilings = 8     # should be >= 4*dim
dim = 2
tile_length = 1.0 / tiles
### ------------------------ ###
```

```python
# PyFixedReps.TileCoder's cascade
offsets_cascade = []
for n in range(tilings):
    offsets_cascade.append(np.ones(dim) * n * (tile_length / tilings))

plot_grey_squares(offsets_cascade)
```

```python
# PyFixedReps.TileCoder's even
offsets_even = []
for n in range(tilings):
    i = n - (tilings / 2)
    offsets_even.append(np.ones(dim) * i * (tile_length / tilings))

plot_grey_squares(offsets_even)
```

```python
# RL textbook's asymmetric tiling
offsets_asym = []
for n in range(tilings):
    offset = []
    for d in range(dim):
        offset.append(((1 + 2*d) * n * (tile_length / tilings)) % tile_length - tile_length / 2)
    offsets_asym.append(offset)

plot_grey_squares(offsets_asym)
```
