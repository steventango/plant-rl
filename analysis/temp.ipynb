```python
import numpy as np
import matplotlib.pyplot as plt
from utils.metrics import UnbiasedExponentialMovingAverage as uema
```

```python
policy1 = np.ones(10)
policy2 = np.array([0,0,1,1,3,3,1,1,0,0])
policy3 = np.array([0,0,0.5,0.5,1,1,1.5,1.5,2,2])

trace1 = uema(alpha=0.1)
trace2 = uema(alpha=0.1)
trace3 = uema(alpha=0.1)
curve1 = []
curve2 = []
curve3 = []
for j in range(1):
    for i in range(len(policy1)):
        trace1.update(policy1[i])
        trace2.update(policy2[i])
        trace3.update(policy3[i])
        curve1.append(trace1.compute().item())
        curve2.append(trace2.compute().item())
        curve3.append(trace3.compute().item())


plt.plot(np.array([curve1, curve2, curve3]).T)
```
